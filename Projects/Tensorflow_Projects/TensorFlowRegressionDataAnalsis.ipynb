```python
X=housedata.drop('price',axis=1).values
y=housedata['price'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
# here we are just tranforming the train data and test data using MinMaxScaler Object 

X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)


```

# here we are going to perfomr tensorflow keras model

```python
#here we are import  the sequential using the keras model using tensorflow
# then also importing the dense using the layers using tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

```

```python
X_train.shape # here below we could see 22 thats the neauron layer 
```

```python
#we will create the object of the sequential
#and also now we will train our data also adding dense layerr
```

```python
model=Sequential()
model.add(Dense(22,activation='relu'))
model.add(Dense(22,activation='relu'))
model.add(Dense(22,activation='relu'))
model.add(Dense(22,activation='relu'))

#final one of adding the dense layer
model.add(Dense(1))


model.compile(optimizer='adam',loss='mse')

```

# now we will fit our data model which means we will train our data as well as validate also


```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),
         batch_size=128,epochs=400) # here beloe we trained our data
```

# if we want to see loss of our model data you could use history to get that will return to the dict

```python
model.history.history
```

```python
pd.DataFrame(model.history.history) #In below we could see the loss as well as the val_loss also that is our test_data loss
# we can see the loss on trainig as well loss on testing
```

```python
testndtrainloss=pd.DataFrame(model.history.history)
testndtrainloss.plot()
```

# we will find the mean absolute error

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
prediction_price=model.predict(X_test)
```

```python
print(np.sqrt(mean_squared_error(y_test,prediction_price)))
```

```python
print(mean_absolute_error(y_test,prediction_price))
```
