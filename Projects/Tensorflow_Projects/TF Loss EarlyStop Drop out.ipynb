```python
#wheneve n we want to perform any classification  first thing we need the values only for 0 or 1 this is the dependent variable

X=cancerdata.drop('benign_0__mal_1',axis=1).values
y=cancerdata.benign_0__mal_1.values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
#now we want to pre process the data using minmaxscalar
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar=MinMaxScaler()
```

```python
X_train=scalar.fit_transform(X_train)
```

```python
X_test=scalar.transform(X_test)
```

```python
#one thing you should know if we want to train our model data we can do for sklearn lib package if we want seee the prediction model
#and loss data and all we can use whatever like algorithm and tensorFlow we can predict see out loss data like how our algorithm predict wheather its correct or not
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(30,activation='relu')) # 30 is the nurons relu means rectifiying linear unit

# here we will add multiple layers and more nurons
model.add(Dense(15,activation='relu'))
# when the last neuron layer is always one 
# and note things if in case we are perform thing the binary classfication that time we muse use the AF is sigmoid because this 
#binary classification

model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
#we will train our data
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))# here when we give large num of epoch then we can see the overfit data
```

```python
#now above we trained our data
model_loss=pd.DataFrame(model.history.history)
model_loss
```

```python
model_loss.plot() # here below plot we can see the val_loss which means our prediction loss its totally over fit

# we need to predict because we already validate our test duing  the  training time above stpes
```
